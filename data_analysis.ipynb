# ECS 171 Bop or Flop: Data Analysis
Using exploratory data analysis to find the correlation of data and to understand the data distribution.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.figure_factory as ff


database = pd.read_csv('SpotifySongPolularityAPIExtract.csv')
database.head()
```

### Distribution of All Numeric Variables

```python
import math
print('\033[1mNumeric Features Distribution'.center(100))

n = 5
cols = database.columns.tolist()[3:]
n_col = len(cols)

clr=['r','g','b','g','b','r']

plt.figure(
    figsize=[15,4*math.ceil(n_col/n)])

for i in range(n_col):
    plt.subplot(math.ceil(n_col/3),n,i+1)
    sns.distplot(database[cols[i]],hist_kws=dict(edgecolor="black", linewidth=2), bins=10, color=list(np.random.randint([255,255,255])/255))

plt.tight_layout()
plt.show()
```

### Correlation Matrix

```python
# Generate the plot
# take out String columns
data = database.copy().drop(columns=['artist_name', 'track_name', 'track_id']).corr()
pcmp = sns.heatmap(data, vmin=-1, vmax=1, center=0, cmap= 'RdBu')
```

### Pairplot (Scatter Plot Matrix)

```python
# Generate a pairplot (scatter plot matrix)
fig = ff.create_scatterplotmatrix(database, diag='histogram', index='artist_name', opacity=.8)
fig.update_layout(
    autosize=False,
    width=800,
    height=800,
)
fig.show()
```
